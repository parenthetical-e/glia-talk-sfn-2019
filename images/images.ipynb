```python
import os
import numbers
import numpy as np
from scipy.stats import norm
from collections import defaultdict

from IPython.display import Image
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

sns.set(font_scale=2)
sns.set_style('ticks')
matplotlib.rcParams.update({'font.size': 16})
matplotlib.rc('axes', titlesize=16)
```

# Gliotransmission

```python
xs = np.linspace(-3, 3, 100)
ys = norm.pdf(xs)
```

```python
fig = plt.figure(figsize=(1,2))
grid = plt.GridSpec(1, 1, wspace=0.8, hspace=1)
plt.subplot(grid[0, 0])
plt.plot(xs, ys, c="grey", linewidth=4)
plt.fill_between(xs, ys, color="grey")
plt.axis('off')
sns.despine()
```

```python
fig = plt.figure(figsize=(0.5,1))
grid = plt.GridSpec(1, 1, wspace=0.8, hspace=1)
plt.subplot(grid[0, 0])
plt.plot(xs, ys, c="grey", linewidth=4)
plt.fill_between(xs, ys, color="grey")
plt.axis('off')
sns.despine()
```

# Phi

```python
def phi(xs, beta=1, b=-2):
    return (1 / 1 + np.exp(xs/beta)) + b
```

```python
xs = np.linspace(0, 50, 100)
ys = phi(xs, beta=1, b=0)
```

```python
fig = plt.figure(figsize=(8,2))
grid = plt.GridSpec(1, 2, wspace=0.8, hspace=1)
plt.subplot(grid[0, 0])
plt.plot(xs, ys, c="grey", linewidth=4)
plt.ylim(0, 20)
plt.xlim(0, 4)
plt.xlabel("I (namps)", fontname="PT Sans")
plt.ylabel("Firing rate (spk/s)", fontname="PT Sans")
sns.despine()

plt.subplot(grid[0, 1])
plt.plot(xs, ys, c="grey", linewidth=4)
plt.ylim(0, 20)
plt.xlim(0, 4)
plt.xlabel("Pixel value", fontname="PT Sans")
plt.ylabel("Firing rate", fontname="PT Sans")
sns.despine()

plt.savefig("phi.png", dpi=fig.dpi,  bbox_inches='tight')
```
